## Ahora solucionaremos el problema de los datos faltantes (nulos o NaN). 

<div class="alert alert-success">
  Los datos faltantes generan problemas con muchos algoritmos de ML. Es por esto que existen distintas estrategias para lidiar con ellos.
</div>

```python
help(pd.Series.value_counts)
```

```python
movies_v2.isna().sum(axis=0)#/len(movies_v2)       #columnas con mas datos faltantes
```

```python
movies_v2.notnull().apply(pd.Series.value_counts)     #apply viene siendo un for loop de la columna x columna, es otra forma de representar los datos faltantes como arriba 
```

```python
(movies_v2 != 0).apply(pd.Series.value_counts)        #Se buscan lo valores con ceros, para determinar si tienen un impacto en el estudio 
```

```python
available = ((movies_v2 != 0) & (movies_v2.notnull()))        #Creamos una boolean filtering de todo el df, para quitar valores nulos y en 0 de nuestro target, porque no podemos aplicar el promedio o la mediana sobre estos 
```

```python
available.head()        #Creamos una boolean filtering
```

```python
available.all(axis=1).value_counts()   #Contar los falsos del df anterior 
```

No podemos entrenar nuestro algoritmo con datos cuya variable objetivo no esta definida o sea nula (valor falso). Eliminemos esas líneas.

```python
mask = available['worldwide_gross']         #Creamos la mascara de worldwide_gross para poder quitar valores 0s
mask
```

```python
movies_v2 = movies_v2[mask]                    #Mandamos pedir los trues de la mascara sobre el df, ya no trae valores 0s en nuestro target 
movies_v2.head()
```

```python
((movies_v2 != 0) & (movies_v2.notnull())).worldwide_gross.value_counts()    #Contamos los valores nulos y 0s en nuestro target, si esta tod en true ya esta limpia 
```

```python
movies_v2 = movies_v2.drop('movie_title',axis=1)   #Quitamos las columnas identificadoras
```

```python
movies_v2 = movies_v2.drop('duration',axis=1)   #Quitamos columnas repetidas
```

```python
movies_v2 = movies_v2.drop('gross',axis=1)    #Sabemos que se tienen que quitar estas variables pero para caso practico las mantenemos
```

```python
#movies_v2 = movies_v2.drop('budget',axis=1)
```

```python
movies_v2.head()
```

```python
movies_v2 = movies_v2[available.screens]    #SeEstan quitando los screens por medio de mascara para quitar valores nulos
movies_v2.head()
```

```python
len(movies_v2)
```

<div class="alert alert-success">
  En el caso de las features que no son la variable objetivo una mejor solución para lidiar con los datos faltantes es <strong>remplazar estos datos por otros que sean manejables</strong> y no afecten la calidad de las predicciones. La estrategia más comun es utilizar la media de todos los ejemplos para la features dadas.
</div>

```python
from sklearn.impute import SimpleImputer     # Se reemplazan los valores nulos de las columnas, con valores promedio o mediana
imputer = SimpleImputer(missing_values= np.nan,  strategy = 'mean')
```

```python
values = imputer.fit_transform(movies_v2)   #Llenamos los valores nulos con el promedio 
X = pd.DataFrame(values)       #convertimos nuestros valores en un df 
X.columns = movies_v2.columns    #Y le pasamos los nombres de las columnas 
X.index = movies_v2.index         #Y el indice 
X.head()
```

```python
len(X)
```

```python
movies_v2.values     #Diferencias entre los valores en el df con datos nulos
```
